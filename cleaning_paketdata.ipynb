```python
df_data = pd.read_csv("paket_data.csv", encoding = 'unicode_escape')
df_data.head(10)
```

```python
df_data['Name'] = df_data['Name'].str.lower()
df_data['Description'] = df_data['Description'].str.lower()
df_data['Operator'] = df_data['Operator'].str.lower()
```

```python
df_data.head(5)
```

```python
name = []
kuota = []
durasi = []
```

```python
x = 0
for i in range(len(df_data['Name'])):
  data = df_data['Name'][i].replace(',','')
  data = data.replace(' gb','gb')
  data = data.replace(' mb','mb')
  data = data.replace(' hari','hr')
  data = data.replace('hari','hr')
  data = data.replace('/',' ')
  split_data = data.split()
  found = 0
  for j in range(len(split_data)):
    if 'gb' in split_data[j]:
      x+=1
      nama_paket = split_data[:j]
      listToStr = ' '.join(str(elem) for elem in nama_paket)
      name.append(listToStr)
      kuota.append(split_data[j])
      if 'hr' in split_data[len(split_data)-1]:
        durasi.append(split_data[len(split_data)-1])
      elif 'bulan' in split_data[len(split_data)-1]:
        time = split_data[len(split_data)-2]*30
        durasi.append(time)
      else:
        durasi.append("0")
      found = 1
      break
    elif split_data[j][-2:] == 'mb':
      x+=1
      nama_paket = split_data[:j]
      listToStr = ' '.join(str(elem) for elem in nama_paket)
      name.append(listToStr)
      kuota.append(split_data[j])
      if 'hr' in split_data[len(split_data)-1]:
        durasi.append(split_data[len(split_data)-1])
      elif 'bulan' in split_data[len(split_data)-1]:
        time = split_data[len(split_data)-2]*30
        durasi.append(time)
      else:
        durasi.append('0')
      found = 1
      break

  if found == 0:
    kuota.append('')
    if 'hr' in split_data[len(split_data)-1]:
      nama_paket = split_data[:len(split_data)-1]
      listToStr = ' '.join(str(elem) for elem in nama_paket)
      name.append(listToStr)
      durasi.append(split_data[len(split_data)-1])
    elif 'bulan' in split_data[len(split_data)-1]:
      nama_paket = split_data[:len(split_data)-1]
      listToStr = ' '.join(str(elem) for elem in nama_paket)
      name.append(listToStr)
      time = split_data[len(split_data)-2]*30
      durasi.append(time)
    else:
      name.append(data)
      durasi.append('0')
    x+=1
print(x)
```
